```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
df = pd.read_csv('movie_booking.csv')
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe(include='all')
```

```python
df[['occu_perc', 'capacity']] = df[['occu_perc', 'capacity']].fillna(df[['occu_perc', 'capacity']].mean())
```

```python
df.isnull().sum()
```

```python
df= df.drop_duplicates()
```

```python
#total sales per date
plt.style.use('fivethirtyeight')
df.groupby('date')['total_sales'].sum().plot(figsize=(16, 4), color='skyblue')
```

```python
#Total sale per day cinema
plt.style.use('fivethirtyeight')
df.groupby('cinema_code')['total_sales'].sum().plot(figsize=(22, 4), color='skyblue')
```

```python
plt.style.use('fivethirtyeight')
df.groupby('film_code')['total_sales'].sum().plot(figsize=(18, 4), color='skyblue')
```

```python
plt.style.use('fivethirtyeight')
cg = df.groupby('cinema_code', as_index=False)['total_sales'].sum()
cg = cg[cg.total_sales > 9.000000e+09]
plt.figure(figsize=(24, 5))
sns.barplot(data=cg, x='cinema_code', y='total_sales', palette='autumn_r')
```

```python
plt.style.use('fivethirtyeight')
fg = df.groupby('film_code', as_index=False)['total_sales'].sum()
fg = fg[fg.total_sales > 9.000000e+09]
plt.figure(figsize=(22, 6))
sns.barplot(data=fg, x='film_code', y='total_sales', palette='prism_r')
```
